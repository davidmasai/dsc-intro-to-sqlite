```python
import sqlite3

# Create a connection to the SQLite database
new = sqlite3.connect('data.sqlite')  # `new` is the connection object

# Create a cursor object to interact with the database
all_records = new.cursor()  # `all_records` is the cursor object

# Execute an SQL query to select all records from the 'offices' table
all_records.execute("""
    SELECT *
    FROM offices;
""")

# Fetch 5 rows from the result set
rows = all_records.fetchmany(5)

# Print the fetched rows
for row in rows:
    print(row)

# # Don't forget to close the connection after you're done
# new.close()
```
